#### Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?

```python
# Given data
sample_mean = 200  # Sample mean
sample_std = 30  # Sample standard deviation
sample_size = 2000  # Sample size
population_size = 3000000  # Population size

# Degrees of freedom (for calculating the z-score)
degrees_of_freedom = sample_size - 1

# Calculate the standard error (standard deviation of the sample mean)
standard_error = sample_std / (sample_size**0.5)

# Calculate the z-score for different confidence levels
confidence_levels = [0.94, 0.98, 0.96]
confidence_intervals = {}

for level in confidence_levels:
    # Calculate the z-score based on the confidence level
    z_score = st.norm.ppf(1 - (1 - level) / 2)

    # Calculate the margin of error
    margin_of_error = z_score * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    confidence_intervals[f'{int(level * 100)}% Confidence Interval'] = (lower_bound, upper_bound)

# Print the confidence intervals
for interval, bounds in confidence_intervals.items():
    print(f"{interval}: {bounds}")
```

#### Q13) What is the nature of skewness when mean, median of data are equal?
- When the mean and median of a dataset are equal, it generally indicates that the distribution of the data is symmetric. In such cases, the data distribution tends to have a symmetrical shape around the center, where the mean and median lie.

- However, it's important to note that a dataset having equal mean and median doesn't necessarily imply that the distribution is perfectly symmetric. It's possible to have symmetric distributions where the mean and median are equal, but there might be various degrees of skewness present in the distribution that cancel out each other, resulting in the equality of mean and median.


#### Q14) What is the nature of skewness when mean > median ?


When the mean of a dataset is greater than the median, it suggests that the distribution of the data is right-skewed or positively skewed.

A right-skewed distribution means that the tail of the distribution extends towards higher values or the right side of the median, while the majority of the data points are concentrated towards the lower values or the left side of the median. This elongation towards the higher values pulls the mean towards that direction, making it greater than the median.

In a right-skewed distribution:
- The mean is greater than the median.
- The tail of the distribution is longer on the right side.
- The majority of values are concentrated on the left side, and fewer values extend towards higher values.

This type of skewness is commonly observed in distributions like income levels (where a few individuals have significantly high incomes), housing prices, and many other economic or financial datasets.

It's important to note that the relationship between the mean and median, such as mean being greater than median, is an indication of skewness but might not provide a complete understanding of the distribution's shape. Visual inspection of a histogram or computing skewness directly can provide a more comprehensive understanding of the skewness in the dataset.


#### Q15) What is the nature of skewness when median > mean?


When the median of a dataset is greater than the mean, it indicates that the distribution of the data is left-skewed or negatively skewed.

A left-skewed distribution implies that the tail of the distribution extends towards lower values or the left side of the median, while the majority of the data points are concentrated towards higher values or the right side of the median. This elongation towards the lower values pulls the mean towards that direction, making it smaller than the median.

In a left-skewed distribution:
- The median is greater than the mean.
- The tail of the distribution is longer on the left side.
- Most of the values are concentrated on the right side, and fewer values extend towards lower values.

Left-skewed distributions are commonly observed in scenarios like the lifespan of products (where there's a minimum but no maximum limit), test scores (where the minimum score could be 0, but there's no maximum limit), and similar datasets where there's a natural limit on one end of the values.

As always, while the relationship between mean and median (median being greater than mean) is an indication of skewness, further analysis, such as visual inspection or computing the skewness directly, can offer a more comprehensive understanding of the shape of the distribution.

#### Q16) What does positive kurtosis value indicates for a data ?


A positive kurtosis value in a dataset indicates that the distribution has heavier tails and a more peaked or sharper central peak (leptokurtic distribution) compared to a normal distribution. It signifies that the dataset has more extreme values (outliers) in the tails than would be expected under a normal distribution.

In a distribution with positive kurtosis:
- The tails of the distribution are heavier than those of a normal distribution.
- There is an increased probability of extreme values, making the distribution more prone to outliers.
- The central peak of the distribution is higher and sharper than that of a normal distribution.

Positive kurtosis doesn't necessarily imply the direction of skewness. A distribution can have positive kurtosis and be either left-skewed, right-skewed, or approximately symmetric.



#### Q17) What does negative kurtosis value indicates for a data?


A negative kurtosis value in a dataset indicates that the distribution has lighter tails and a flatter peak (platykurtic distribution) compared to a normal distribution. It signifies that the dataset has fewer extreme values (outliers) in the tails than would be expected under a normal distribution.

In a distribution with negative kurtosis:
- The tails of the distribution are lighter than those of a normal distribution.
- There is a reduced probability of extreme values, making the distribution less prone to outliers.
- The central peak of the distribution is flatter and lower compared to that of a normal distribution.

Negative kurtosis doesn't specify the direction of skewness. A distribution can have negative kurtosis and be either left-skewed, right-skewed, or approximately symmetric.

 ##### Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days
Hint:  
   rcode  🡪 pt(tscore,df)  
 df 🡪 degrees of freedom

```python
import math
from scipy.stats import norm

population_mean = 270
sample_mean = 260
sample_std = 90
sample_size = 18

# Calculate the standard error (SE)
SE = sample_std / math.sqrt(sample_size)

# Calculate the z-score
z_score = (sample_mean - population_mean) / SE

# Calculate the probability using the cumulative distribution function (CDF) of the standard normal distribution
probability = norm.cdf(z_score)

print(f"The probability that 18 bulbs would have an average life of no more than 260 days: {probability:.2f}")
```
